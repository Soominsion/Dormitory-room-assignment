```python
%%writefile Student.py
class Person():
    def __init__(self,name,birth,gender,adress):
        self._name=name
        self._birth=birth
        self._gender=gender
        self._adress=adress
              
    @property
    def name(self):
        return self._name
    
    @property
    def birth(self):
        return self._birth
    

    @property
    def gender(self):
        return self._gender
    
    @property
    def adress(self):
        return self._adress
    
    
    
        
class Student(Person):
    def __init__(self,name,birth,gender,adress,std_num,std_scr,yes_resi,yes_smo,good_point,sensitive,num):
        Person.__init__(name,birth,gender,adress)
        self._std_num=std_num
        self._std_scr=std_scr
        self._yes_resi=yes_resi
        self._yes_smo=yes_smo
        self._good_point=good_point
        self._sensitive=sensitive


        
    @property
    def std_num(self):
        return self._std_num
    
    
    @property
    def std_scr(self):
        return self._std_scr
    
 
    @property
    def yes_resi(self):
        return self._yes_resi
    
    @property
    def yes_smo(self):
        return self._yes_smo
    
    @property
    def good_point(self):
        return self._good_point
    
    
    @property
    def sensitive(self):
        return self._sensitive
    

    
        
```

```python
%%writefile RoomAssignment.py
from Student import *
from SelectSystem import *

class RoomAssignment():
    
    def __init__(self):
        self.students=[]
        self.grouped_students=[]
    
    def register_student(self,student):
        self.students.append(student)
    
    def process(self):
        
        sorted_students = sorted(self.students, key=lambda x: (x['흡연여부'], x['예민도']))
        self.grouped_students = [sorted_students[i:i+4] for i in range(0, len(sorted_students), 4)]
     
    def print_students(self):
        for i, group in enumerate(self.grouped_students):
            print(f"그룹 {i+1}:")
            for student in group:
                print(student)
            print()               
        
        
        
```

```python
from Student import *
from SelectSystem import *
from RoomAssignment import *
import time

def info_input():
    print("정보 입력시 신중하게 기입하여 주십시오 .")
    print("잘못된 정보가 입력된 경우 입사시 불이익이 발생할 수 있습니다.\n")
    

    name=input("이름을 입력하시오:")
    
    birth=input("생일을 입력하시오(ex.20011231):")
    if len(birth)>8 or int(birth[4:6])>13 or int(birth[6:8])>31 :
        raise ValueError("생년월일 8자리를 \"년도,월,일\" 순으로 정확하게 기입하시오")
        
    
    gender=input("성별을 입력하시오(남/여):") 
    if gender!='남' and gender!='여':
        raise ValueError("\'남\' 또는 \'여\' 로 정확하게 기입하시오")
        
    adress=input("도로명 주소를 입력하시오:") 

    std_num=input("학번을 입력하시오:")
    if len(std_num)>8:
        myerror=ValueError("학번 8자리를 정확하게 기입하시오")
        raise myerror

    std_scr=float(input("학점을 입력하시오:"))
    if std_scr<0 or std_scr>4.3:
        myerror=ValueError("올바르지 않은 숫자입니다.학점을 정확하게 기입하시오 ")
        raise myerror
    
    sensitive=int(input("자신의 예민도를 1~10 까지 정수로 수치화하여 입력하시오(예민할수록 10이라 판단):")) 
    if sensitive>10:
        myerror=ValueError("예민도를 1~10 까지 수치화하여 정확하게 입력하시오")
        raise myerror
        
    yes_smo=input("흡연 여부를 입력하시오(네/아니오):") 
    if yes_smo!='네'and yes_smo!='아니오':
        myerror=ValueError("\'네\' 또는 \'아니오\' 로 정확하게 기입하시오")
        raise myerror
    
    yes_resi=input("기존 거주 여부를 입력하시오(네/아니오):") 
    if yes_resi!='네'and yes_resi!='아니오':
        myerror=ValueError("\'네\' 또는 \'아니오\' 로 정확하게 기입하시오")
        raise myerror
    if std_num[0:4]=='2023':
        myerror=ValueError("1학년은 기존 사생일 수 없습니다")
        raise myerror
    
    if yes_resi=='네':
        good_point=int(input("상점을 입력하시오:") )
    else:
        good_point=0
        
    with open("Dor_student.txt", "ab") as fp:
        pickle.dump({"이름": name,"생일": int(birth),"성별": gender, "주소": adress,"학번": std_num,"학점": float(std_scr),"예민도": int(sensitive),"흡연여부": yes_smo,"기존거주여부": yes_resi,"상점": int(good_point) }, fp)
    
    
    
def calltime(func):  
    def new_func(self):
        start_time=time.time()
        result=func()
        end_time=time.time()
        pro_time=end_time-start_time
        print("실행시간은:",pro_time)
        return result
    return new_func  



class MyFunction:
    @calltime
    def my_function():
        selecting=SelectSystem()
        m_assigmenting=RoomAssignment()
        f_assigmenting=RoomAssignment()
        students = []
        sel_m_std=[]
        sel_f_std=[]

        #학생정보 입력받음
        info_input()
        print("접수가 마감되었습니다.")

        with open("Dor_student.txt", "rb") as fp:
            while True:
                try:
                    students.append(pickle.load(fp))
                except EOFError:
                    break
                    
                    
        
        #학생 선발 과정
        for student in students:
            selecting.register_student(student)
        selecting.process()
        sel_m_std,sel_f_std = selecting.register_selc_student(student)


        #남학생 방배정 과정   
        for student in sel_m_std:
            m_assigmenting.register_student(student)

        m_assigmenting.process() 


        #여학생 방배정 과정   
        for student in sel_f_std:
            f_assigmenting.register_student(student)

        f_assigmenting.process()  
        
        
        #결과출력
        m_assigmenting.print_students()
        f_assigmenting.print_students()
    
    
my_func = MyFunction()
my_func.my_function()
```
